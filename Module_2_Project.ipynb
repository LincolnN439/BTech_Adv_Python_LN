```python
Q3_df = df.groupby('YearBuilt')['Price'].mean().reset_index()

sns.lineplot(x='YearBuilt', y='Price', data=Q3_df)
plt.show()
```

```python
# Q6: According to the bar chart above, properties appear to have the highest average price in which region?
```

```python
df
```

```python
Q5_df = df.groupby('Regionname')['Price'].mean().reset_index()

sns.barplot( x='Regionname', y='Price', data=Q5_df )
plt.xticks(rotation=90)
plt.show()
```
