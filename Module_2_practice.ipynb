```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('companies.csv') 
```

```python
# Q1: Using matplotlib/seaborn, create a countplot that shows a count of how many of the companies in the data set are from each state. 
# Paste a picture of the graph below. 
# Note that you may need to adjust the width/height of the graph appropriately.
```

```python
# Q2: Which state are most of the companies from?
```

```python
SortedDF = df['state'].value_counts().sort_values(ascending = False)
```

```python
sns.countplot(x= 'state', data=df , order=SortedDF.index)
plt.xticks(rotation=90)
plt.figure(figsize=( 10,6 ))
plt.show()

```

```python
# Q3: Using matplotlib/seaborn, create a bar chart that shows a sum of how many total job openings there are in each industry.
# Paste a picture of the graph below. 
# Note that you may need to adjust the width/height of the graph appropriately. 
```

```python
# Q4: Which industy has the most job openings?
```

```python
openings_df= df.groupby('Industry')['job_openings'].sum().reset_index().sort_values(by= 'job_openings')
sns.barplot( y='job_openings' , x='Industry'  , data=openings_df)

plt.xticks(rotation=90)
plt.show()
```
